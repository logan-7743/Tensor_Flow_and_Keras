```python
import tensorflow as tf
import tensorflow_datasets as tfds
```

```python
imbd, info = tfds.load("imdb_reviews",with_info=True,as_supervised=True)
```

```python
imbd_test, imbd_train = imbd["test"],imbd["train"]
```

```python
training_sentences = []
training_labels    = []

test_sentences     = []
test_labels        = []

#Extract data from the tensor data set
for s,l in imbd_test:
  test_sentences.append(s.numpy().decode('utf8'))
  test_labels.append(l.numpy())

for s,l in imbd_train:
  training_sentences.append(s.numpy().decode('utf8'))
  training_labels.append(l.numpy())
```

```python
#Hyper Prarms
vocab_size    = 10000   #How many words to keep track of, keep by highest volumn if we exceed the length
embedding_dim = 16      #Length of embding vector. 16d vector
max_length    = 120     #Make words of review.. maybe
trunc_type    = "post"  #Where to cut off the excess words,
oov_tok       = "<OOV>" #Out of vocab, like a place holder for words we dont know about
```

```python
#Dependencies
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
#Intialize tokenizer
tokenizer = Tokenizer(num_words = vocab_size, oov_token = oov_tok)
#fit our tokenizer to our training reviews
tokenizer.fit_on_texts(training_sentences)
#Get the dictonary
word_index = tokenizer.word_index
#Map word_index to our orginal reviews
sequences = tokenizer.texts_to_sequences(training_sentences)
#Fit the sequences to have same length
padded = pad_sequences(sequences,maxlen=max_length,truncating=trunc_type)
#map word_index to test data
testing_sequences = tokenizer.texts_to_sequences(test_sentences)
#Fit the data to have same size
test_padded = pad_sequences(testing_sequences,maxlen=max_length)
```

```python
#Make nueral net
#Goal for embedding is to find a true relshionship for the words. It uses a vector for the word and tries to find the best fit of each vector
#in predicting something such as next word or surrounding word
#After embeding, we nd up with a bunch of vectors that represent words, so we should flatten.
#Then just have a relu layer, and the output layer being activated by sigmoid.
#The our put of thid model is to predict the review labels
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size,embedding_dim,input_length = max_length),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(6, activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
model.summary()
```

```python
import numpy as np
model.fit(padded,np.array(training_labels),epochs=5,validation_data=(test_padded,np.array(test_labels)))
```
